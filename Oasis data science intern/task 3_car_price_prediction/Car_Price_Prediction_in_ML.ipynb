### **Model Training**

```python
predict = "price"
data = data[["symboling", "wheelbase", "carlength","carwidth", "carheight", "curbweight","enginesize", "boreratio", "stroke","compressionratio", "horsepower", "peakrpm", 
             "citympg", "highwaympg", "price"]]
x = np.array(data.drop([predict], 1))
y = np.array(data[predict])
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25)
```

```python
#Decision tree regressor
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
k = model.fit(xtrain, ytrain)
predictions = model.predict(xtest)
predictions
```

```python
from sklearn.metrics import mean_absolute_error
model.score(xtest, predictions)
```

```python
#RandomForest regressor
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor()
k = model2.fit(xtrain, ytrain)
predictions2 = model2.predict(xtest)
predictions2
```

```python
from sklearn.metrics import mean_absolute_error
model2.score(xtest, predictions2)
```

```python
#Extremely Randomized Tree Regressor
from sklearn.tree import ExtraTreeRegressor
from sklearn.ensemble import BaggingRegressor
model3 = ExtraTreeRegressor()
k = model3.fit(xtrain, ytrain)
predictions3 = model3.predict(xtest)
predictions3
```

```python
from sklearn.metrics import mean_absolute_error
model3.score(xtest, predictions3)
```

```python
model4 = BaggingRegressor(model3, random_state=0).fit(xtrain, ytrain)
predictions4 = model4.predict(xtest)
predictions4
```

```python
from sklearn.metrics import mean_absolute_error
model4.score(xtest, predictions4)
```

#### *All four models which we implemented giving same accuracy i.e 100%*
